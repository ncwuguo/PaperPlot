```python
X = 0.63
Y = 70
e = 300
U = 20000
mu = 10000
h0 = 10
h1 = 5
hg = 12
c0 = 25
c1 = 30
g0 = 20
g = 25
v = 8
s = 5
w = 150
M = 400
o = 10

Qg = U * (1 - (w + hg - v) / (X * (e - v)))
qm = U * (1 - (c1 + h0 + g0 - v - o) / (X * (Y + g - v))) - Qg
qs = U * (1 - (c0 + h1 - s) / (X * (Y - s))) - qm - Qg
Qa = U * (1 - (w + hg - v) / (X * (e - v)))
qa = U * (1 - (c1 + g0 + h0 - v) / (X * (e - v))) - Qa

# 给出高度Z的值
Zas = (Qa + qa) * (c1 - c0)
Zs = (
    (c1 - c0) * Qg
    + (c1 - c0) * qm
    + (s + o - c0 - h1) * qs
    + X * (Y - s) * (qs - ((Qg + qm + qs) ** 2 - (Qg + qm) ** 2) / (2 * U))
)
Zam = (
    Qa * (w - c1 - g0)
    + (v - X * v - c1 - g0 - h0) * qa
    + X * (e * qa - (e - v) * (((Qa + qa) ** 2 - Qa**2) / (2 * U)))
)
Zm = (
    (w - c1 - g0) * Qg
    + (v + o - c1 - h0 - g0) * qm
    + X
    * (
        (e - v) * (qm - ((Qg + qm) ** 2 - Qg**2) / (2 * U))
        + (e - Y - g) * (qs - ((Qg + qm + qs) ** 2 - (Qg + qm) ** 2) / (2 * U))
    )
)
Zag = (w + hg - v + X * v) * Qa + X * (
    (-v * (Qa**2)) / (2 * U)
    - e * (((Qa + qa) ** 2 - Qa**2) / (2 * U))
    + e * qa
    + M * (mu - Qa - qa + ((Qa + qa) ** 2) / (2 * U))
)
Zg = (
    Qg * (w + hg - v)
    + X * v * Qg
    + X
    * (
        (-v * (Qg**2)) / (2 * U)
        + e * (qm + qs)
        - e * (((Qg + qm + qs) ** 2 - Qg**2) / (2 * U))
        + M * (mu - Qg - qm - qs + ((Qg + qm + qs) ** 2) / (2 * U))
    )
)
```
